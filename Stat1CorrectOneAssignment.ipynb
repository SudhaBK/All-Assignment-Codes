### Assignment 1: Basic Statistics "Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset"

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Task 1: Load dataset and calculate statistics

df = pd.read_csv("sales_data_with_discounts.csv")

```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical Columns:", numerical_cols)


# Interpretation
print("The statistics suggest that the data has varying means, medians, and standard deviations.")

stats = df[numerical_cols].describe()
print("Statistics:\n", stats)
```

### Task 2 : Data Visualization : Histogram , Boxplots and Bar Graph

```python

## Histograms
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()

```

```python
## Boxplots
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

```

```python
## Bar Chart Analysis for Categorical Column
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(df[col])
    plt.title(f"Bar Chart of {col}")
    plt.show()
```

### # Task 3: Standardization of Numerical Variables

```python

## Explanation
print("Standardization (z-score normalization) scales data to have a mean of 0 and standard deviation of 1.")

## Standardization
scaler = StandardScaler()
df_std = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)

## Comparison
print("Before Standardization:\n", df[numerical_cols].head())
print("After Standardization:\n", df_std.head())


```

### Task 4: Conversion of Categorical Data into Dummy Variables

```python

## Explanation
print("One-hot encoding converts categorical data into binary columns for each category.")

## One-hot Encoding
df_dummy = pd.get_dummies(df, columns=categorical_cols)
print("Transformed Dataset:\n", df_dummy.head())


```
